# Random tensors

Random tensor are tensors of some abitrary size which contain random numbers

```python
random_1 = tf.random.Generator.from_seed(42)
random_1 = random_1.normal(shape=(3, 2))
random_1
```

```python
random_2 = tf.random.Generator.from_seed(42)
random_2 = random_2.normal(shape = (3, 2))
random_2
```

```python
random_1, random_2, random_1 == random_2
```

```python
random_3 = tf.random.Generator.from_seed(42)
random_3 = random_3.normal(shape = (3,2))
random_3
```

```python
shuffle = tf.random.shuffle(random_3)
random_3, shuffle
```

```python
random_4 = tf.random.Generator.from_seed(42)
random_4 = random_4.uniform(shape = (3, 2))
shuffle_2 = tf.random.shuffle(random_4)
random_4, shuffle_2
```

```python
tf.random.set_seed(42)
not_shuffled = tf.Variable([[10, 7],
                            [3, 4],
                            [1, 2]])
shuffled = tf.random.shuffle(not_shuffled)
not_shuffled, shuffled
```

```python
shuffled.ndim
```

```python
tf.random.set_seed(42)
not_shuffled = tf.constant([10, 2, 3, 4])
shuffled = tf.random.shuffle(not_shuffled)
not_shuffled, shuffled
```

* shape
* rank
* axis or dim
* size

## Aggregating tensors

```python
exp = tf.Variable([[10, 8],
                   [9, 10]])
exp_1 = exp[0, 0].assign(11)
exp, exp_1
```

```python
# aggregating tensors = compressing big chunks of data in a small single unit or so
```

```python
non_abs = tf.constant([-10, -8])
tf.abs(non_abs)
```

forms of aggregation:
* minimum
* maximum
* mean
* sum of tensor

```python
tensor = tf.constant([[112, 492, 10],
                      [45, 22, 90],
                      [1000, 3, 0]])
tf.reduce_mean(tensor), tf.reduce_min(tensor), tf.reduce_max(tensor), tf.reduce_sum(tensor)
```

```python
E = tf.constant(np.random.randint(0, 100, size = 50))
```

```python
E, E.shape, tf.size(E)
```

```python
tf.cast(E, dtype = tf.float32)
```

```python
tf.math.reduce_std(tf.cast(E, dtype = tf.float32)), tf.math.reduce_variance(tf.cast(E, dtype = tf.float32))
```

### one Hot Encoding
